#  Requests

```python
import requests
import pandas as pd
```

## GET

```python
url = 'https://itunes.apple.com/search'
params = {'media': 'music', 'term': 'manizha', 'limit': 100}  
response = requests.get(url, params=params)
response
```

```python
response.status_code
```

```python
response.text
```

```python
response.json()
```

```python
results = response.json()['results']
results
```

```python
df = pd.DataFrame(results)
df
```

```python
df.artistName.value_counts()
```

```python
help(df.to_excel)
```

```python
df.to_excel('data/iTunes_api.xlsx', sheet_name='Manizha', index_label='id')
```

```python
help(df.to_csv)
```

```python
df.to_csv('data/iTunes_api.csv', sep='\t', index=False)
```

## POST

```python
url = 'https://itunes.apple.com/search'
params = {'media': 'music', 'term': 'manizha', 'limit': 30}  
response = requests.post(url, data=params, timeout=10)
response
```

```python
response.json()
```
